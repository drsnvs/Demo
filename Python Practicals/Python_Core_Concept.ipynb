```python
# format(values) - 1 Simple
o = "Hello My name is {} and i am a {} .how are {} "
o = o.format('Darshan','Student',"you")
print(o)
```

```python
help(str.format)
```

```python
# format(value) - 2 With the uses of index
o = "Hello My name is {0} and i am a {1} .how are {2} "
o = o.format('Darshan','Student',"you")
print(o)
```

```python
# format(value) - 2(2) With the uses of index (keywords)
o = "Hello My name is {2} and i am a {1} .how are {0} "
o = o.format('Darshan','Student',"you")
print(o)
```

```python
# format(value) - 3 With the uses of index name
o = "Hello My name is {name} and i am a {profession} .how are {any} "
o = o.format(name = 'Darshan',profession = 'Student',any = "you")
print(o)
```

```python
# format_map(dictionary_name) : It is used for Dictionary 
help(str.format_map)
x = {"name" : "Darshan",
     "work" : "student"}
print("My name is {name} and i am a {work}.".format_map(x))
```

```python
# if .. elif .. else ..
# Grouped if .. else .. (with the help of logical operators)
# voting

voting_age= 4

if (voting_age>18) and (voting_age<=100):
    print('You are Applicable for voting')
elif (voting_age>0) and (voting_age<18):
    print('Now You are not applicable for voting\nAfter',(18-voting_age),' Year you can vote')
elif(voting_age==18):
    print('After completation of 18 you can vote')
else:
    print('You are not Applicable for voting Please enter Valid Age')
```
