```python
bank = dict()
```

```python
# add
def up():
    ac = 1020
    new_ac = len(bank)+ac
    return new_ac
def upp():
    bank.setdefault(up(),dict())

def wrng_pin():
  print("Wrong Pin Number")
def wrng_ac():
  print("Wrong Account Number")
def msg():
  print("Successfully Completed")
# Formate
def frmt(name,pin,bal):
  z = {
      'name':name,
      'pin':pin,
      'balance':float(bal),
      'transaction':list()
  }
  return z
# Enter Data

# bank.update(frmt('Ritik',5252,0.0))

# Search


def src(user):
  if user in bank:
    # print(bank.get(user))
    for i in bank[user]:
      print(i,bank[user].get(i))

# debit

def wid(ac,amt):
  if ac in bank:
    if bank[ac]["balance"]>amt:
      bank[ac]["balance"]-=amt
      # print(ac,bank[ac]["balance"])
      bank[ac]["transaction"].append(
          {
              'dr_cr':'dr',
              'amount':float(amt)
          }   
      )
    else:
      print("Insufficience Balance")


# credit
def deposit(ac,amt):
  if ac in bank:
    if bank[ac]["balance"]>=0:
      bank[ac]["balance"]+=amt
      # print(ac,bank[ac]["balance"])
      bank[ac]["transaction"].append(
          {
              'dr_cr':'cr',
              'amount':float(amt)
          }
          
      )
    else:
      print("Enter Valid Amount")
  else:
    wrng_ac()
# transaction
def transaction(ac):
  if ac in bank:
    print("Welcome,",bank[ac]["name"],"Your Bank Balance is : ",bank[ac]["balance"],"\nAnd Your Transactions are:")
    tr = bank[ac]["transaction"]
    # print(tr)
    # print(tr)
    # print(tr2)
    ky = bank[ac]["transaction"][0].keys()
    for i in ky:
      print(i,end="\t\t|")
    print()
    dic = bank[ac]["transaction"]
    for i in dic:
      for j in i.values():
        print(j,end="\t\t|")
      print()

def ckbal(ac):
  if ac in bank:
    print("Welcome",bank[ac]["name"],"\nYour Balance Is :",bank[ac]["balance"])


def create_ac():
  name = input(print("Enter Your Name : "))
  pin = input("Enter New pin : ")
  pin2 = input("Enter New pin again : ")
  if pin == pin2:
    bal = input("Deposit Money : ")
    if name.isalnum() and pin.isnumeric() and bal.isnumeric():
      pin = int(pin)
      pin2 = int(pin2)
      bal = float(bal)
      ac_num = up()
      ac_num = int(ac_num)
      upp()
      print("Account Number Is : ",ac_num)
      bank[ac_num].update(frmt(name,pin,bal))
      src(ac_num)
      msg()
    else:
      print("Error")
  else:
    print("Pin Error")

def check_bal():
  acc = int(input("Enter Account Number : "))
  if acc in bank:
    ck = int(input("Enter Your Pin : "))
    if ck == bank[acc]["pin"]:
      ckbal(acc)
      msg()
    else:
      wrng_pin()
  else:
    wrng_ac()

def credit_ac():
  ac = int(input("Enter Your Account Number : "))
  if ac in bank:
    ck = int(input("Enter Your Pin : "))
    if ck == bank[ac]["pin"]:
      dd = float(input("Enter Amount you want to Credit : "))
      deposit(ac,dd)
      msg()
    else:
      wrng_pin()
  else:
    wrng_ac()

def debit_ac():
  cd = int(input("Enter Your Account Number : "))
  if cd in bank:
    ck = int(input("Enter Your Pin : "))
    if ck == bank[cd]["pin"]:
      dd = float(input("Enter Amount you want to Debit : "))
      wid(cd,dd)
      msg()
    else:
      wrng_pin()
  else:
    wrng_ac()

def transaction_ac():
  ac_nm = int(input("Enter Account Number : "))
  if ac_nm in bank:
    pn = int(input("Enter Your Pin Number : "))
    if pn == bank[ac_nm]["pin"]:
      transaction(ac_nm)
      msg()
    else:
      wrng_pin()
  else:
    wrng_ac()
```

```python

choice=int(input(print("Choose a option : \n1.Create Account\n2.Check Balance\n3.Credit\n4.Debit\n5.Transactions : ")))

if choice == 1:
  create_ac()
elif choice == 2:
  check_bal()
elif choice ==3:
  credit_ac()
elif choice==4:
  debit_ac()
elif choice == 5:
  transaction_ac()
else:
  print("Enter Valid Choice")
```
